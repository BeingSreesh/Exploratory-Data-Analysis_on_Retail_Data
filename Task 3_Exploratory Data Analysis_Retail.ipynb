```python
sns.set_palette('magma')
sns.pairplot(df)
plt.show()
```

```python
sns.catplot(x="Category", y="Sales", hue="Category", kind="bar", data=df)

sns.catplot(x="Category", y="Profit", hue="Category", kind="bar", data=df)
```

```python
#quantity of products ordered by each segment

df_group = df.groupby('Segment')['Quantity'].sum().reset_index()
print(df_group)
labels = df['Segment'].unique()
colors = ['green', 'red', 'blue']
plt.figure(figsize=(5,5))
plt.pie(df_group['Quantity'],autopct='%1.1f%%',labels=labels,explode=(0.02,0.02,0.02), colors=colors)
plt.title('Quantities ordered by each segment',size= 20)
plt.show()


df_group2 = df.groupby('Segment')['Profit'].count().reset_index()
print(df_group2)
labels = df['Segment'].unique()
colors = ['green', 'red', 'blue']
plt.figure(figsize=(5,5))
plt.pie(df_group2['Profit'],autopct='%1.1f%%',labels=labels,explode=(0.02,0.02,0.02), colors=colors)
plt.title('Profit % made by each segment',size= 20)
plt.show()
```

```python
#Shipping Models

plt.figure(figsize=(7,4))
sns.countplot('Ship Mode',data=df, palette='rocket_r')
plt.title('Popular Shipping Modes',size=20)
plt.xlabel('\n Shipping mode',size=12)
plt.ylabel('Profit per sale',size=12)
plt.xticks(fontsize=12)
plt.show()
```

```python
#Shipping Mode vs Count

sns.catplot('Ship Mode',data=df,hue='Segment',kind='count',palette='dark',aspect=2,height=6)
plt.title('Segments in Ship modes',size=20)
plt.show()
```

```python
#state wise count for unique values
df['State'].value_counts()
```

```python
#States with least and most ordered

sns.catplot(y="State", kind='count',data=df,palette='Set1',height=7.6,aspect=1.5)
plt.title("Profit Count in different States",size=20)
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 most profit gained cities
dftop10 = df.groupby('City')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
dftop10
print("Top 10 most profit count having cities")
dftop10.head(10)
```

```python
#Top 10 most sales happened cities
dftop10 = df.groupby('City')['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False)
dftop10
print("Sales occurred in 10 cities")
dftop10.head(10)

```

```python
"""
Since the profit count and actual profit amount
showing different countries.
I will find Profit per sale
"""

df['ProPSale'] = df['Profit']/df['Sales']
df.head(4)
```

```python
#product quantities ordered region wise

df_region=df.groupby('Region')['Quantity'].sum().reset_index()
print(df_region)

labels = df_region['Region'].unique()
plt.figure(figsize=(5,5))
plt.pie(df_region['Quantity'],autopct='%1.1f%%', labels=labels, explode=(0.02,0.02,0.02,0.02),)
plt.title('Quantities ordered by each region',size=20)

plt.show()
```

```python
#product profit gained by each region 

df_region=df.groupby('Region')['Profit'].sum().reset_index()
print(df_region)

labels = df_region['Region'].unique()
plt.figure(figsize=(5,5))
plt.pie(df_region['Profit'],autopct='%1.1f%%',labels=labels, explode=(0.02,0.02,0.02,0.02),)
plt.title('Profit gained by each region',size=20)

plt.show()
```

```python
#product profit per sale gained by each region 

df_region=df.groupby('Region')['ProPSale'].count().reset_index()
print(df_region)

labels = df_region['Region'].unique()
plt.figure(figsize=(5,5))
plt.pie(df_region['ProPSale'],autopct='%1.1f%%',labels=labels, explode=(0.02,0.02,0.02,0.02),)
plt.title('Profit per sale gained by each region',size=20)

plt.show()
```

```python
#Highest Selling Categories

df_cats = df.groupby('Category')['Quantity'].sum().reset_index()
df_cats
```

```python
#HIghest Profit per sale gaining category
df_cats = df.groupby('Category')['ProPSale'].sum().reset_index()
df_cats
```

```python
#Highest Profit gaining category
df_cats = df.groupby('Category')['Profit'].sum().reset_index()
df_cats
```

```python
#Highest Sales happening in which category?
df_cats = df.groupby('Category')['Sales'].sum().reset_index()
df_cats
```

From the above results, we can see that Category Technology gaining more profit and also the only category having the most sales but we need to see in which category profit per sale is happening and that is **Office Supplies**

```python
df_cat= df.groupby('Category')['ProPSale'].sum().reset_index()
print(df_cat)
labels = df['Category'].unique()
colors = ['green', 'red', 'blue']
plt.figure(figsize=(5,5))
plt.pie(df_cat['ProPSale'],autopct='%1.1f%%',labels=labels,explode=(0.02,0.02,0.02), colors=colors)
plt.title('Profit per sale % made by each category',size= 20)
plt.show()
```

```python
#Most Profitable Products

dftop10_items = df.groupby('Sub-Category')['ProPSale'].sum().reset_index().sort_values(by='ProPSale',ascending=False)
dftop10_items.reset_index(drop=True,inplace=True)
dftop10_items=dftop10_items.head(10)
dftop10_items
```

```python
#Visualizing the Profit per Sale of Top 10 Profitable Products

sns.catplot('Sub-Category','ProPSale',data=dftop10_items,kind='bar',aspect=1.5,height=9,palette='ch:2.5,-.2,dark=.3')
plt.title('Top 10 profitable products',size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('Profit per Sale',size=18)
plt.xlabel('Products',size=18)

plt.show()
```

```python
#Most Profitable Cities

dftop10_city = df.groupby('City')['ProPSale'].sum().reset_index().sort_values(by='ProPSale',ascending=False)
dftop10_city.reset_index(drop=True,inplace=True)
dftop10_city=dftop10_city.head(10)
dftop10_city
```

```python
sns.catplot('City','ProPSale',data=dftop10_city,kind='bar',aspect=1.5,height=9,palette='ch:2.5,-.2,dark=.3')
plt.title('Top 10 profitable cities',size=20)
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('Profit Per Sale',size=18)
plt.xlabel('Products',size=18)

plt.show()
```
